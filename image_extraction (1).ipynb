```python
import os
import glob
import pandas as pd
import numpy as np
import fitz, sys #pip3 install pymupdf (otherwise the fitz module is not found)
import shutil # to delete directories recursively

# Creating a list of all papers given as input in pdf format under the directory data.
files_path="data/"
read_files= glob.glob(os.path.join(files_path,"*.pdf"))

print('Done1')
```

```python
from PIL import Image
import csv
#print(os.getcwd()) # printing current work directory

grayscale = 1

if grayscale == 0:
    # Removing the Images folder
    if os.path.exists("Images"):
        os.system("rm -rf "+"Images")
    # Creating an empty Images directory
    os.mkdir("Images")
else:
    print("grayscale is 1")
    # Removing the Images_gray folder
    if os.path.exists("Images_gray"):
        os.system("rm -rf "+"Images_gray")
    # Creating an empty Images_gray directory
    os.mkdir("Images_gray")

# for loop to parse through each pdf and extract images from it.
for item in read_files:
  #To open doc using MyPuPDF
    doc = fitz.open(item)
    if grayscale == 0:
        # Item is name of the pdf file which start with the folder name also
        # Ex: data/abc.pdf
        # here we are removing the initial "data/" and the trailing ".pdf" from the name
        # This trimmed name is used to create new diretory/folders.
        dirpath = os.path.join("Images/",item[5:(len(item)-4)])
    else:
        dirpath = os.path.join("Images_gray/",item[5:(len(item)-4)])
    # creating new folder for the current paper.
    os.mkdir(dirpath)
    # changing directory to enter the newly created directory.
    os.chdir(dirpath)
    # Below for loop is to parse the paper for images and store them as .png files
    for i in range(len(doc)):
        for img in doc.getPageImageList(i, full=True):
            xref = img[0]  # check if this xref was handled already?
            pix = fitz.Pixmap(doc, xref)
            # if we want to store the images as rgb (colored)
            if grayscale == 0:
                # If the original image is either grayscale or rgb we store them as is.
                if pix.n < 4:  # this is GRAY or RGB
                    pix.writePNG("p%s-%s.png" % (i, xref))
                # If the original image is CMYK format it first needs to be converted to RBG to store as .png
                # CMYK file cannot be store as .png
                else:  # CMYK needs to be converted to RGB first
                    pix1 = fitz.Pixmap(fitz.csRGB, pix)  # make RGB pixmap copy
                    pix1.writePNG("p%s-%s.png" % (i, xref))
                    pix1 = None  # release storage early (optional)
            # if we want to store the images as grayscale (b/w)
            else:
                # we determine the image is in which format by comparing the number of pixels for each image
                # Grayscale format has only two pixel (black and white)
                # RGB format has three pixel (Red Green and Blue)
                # CMYK has 4 pixels (Cyan, Magenta, Yellow, Key/Black)
                # if the original image is grayscale we store it as is.
                if pix.n < 3:  # this is GRAY
                    pix.writePNG("p%s-%s.png" % (i, xref))
                # if the original image is RGB or CMYK we convert it to grayscale.
                else: # this if for RGB or CMYK
                    pix1 = fitz.Pixmap(fitz.csGRAY, pix)
                    pix1.writePNG("p%s-%s.png" % (i, xref))
                    pix1 = None  # release storage early (optional)
            pix = None  # release storage early (optional)
    
    ################
    #Disread the next few lines they are for some experiment i am runnin and might need later
    ################
    # Getting the list of all images in the current directory.
    #image_list = os.listdir('./')
    #print(image_list)
    #print(os.getcwd()) # printing current work directory
    #for image in image_list:
    #    img_file = Image.open(image)

        # get original image parameters...
    #    width, height = img_file.size
    #    format = img_file.format
    #    mode = img_file.mode

        # Make image Greyscale
    #    img_grey = img_file.convert('L')

        # Save Greyscale values
    #    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    #    value = value.flatten()
    #    print(value)
    #    with open("../../img_pixels.csv", 'a') as f:
    #        writer = csv.writer(f)
    #        writer.writerow(value)
    os.chdir("../../")
print('Done2')
print(os.getcwd())
```

```python
# this section of code is to store the images along with the pdf name in xlsx file, to create a databse of images.
# currently we are storing each image in its original dimensions, 
# do we need to change dimensions of each image to a constant?

import os
import xlsxwriter

if grayscale == 0:
    # entering the image database
    os.chdir("Images")
else:
    # entering the image database
    os.chdir("Images_gray")

# Step 1: Check is the database workbook exists. If yes delete it. We want to start afresh creation of workbook
if os.path.exists("image_database.xlsx"):
    os.system("rm -rf "+"image_database.xlsx")
    
# creating a list of all papers
list_of_papers = os.listdir("./")

# going through each paper and adding the list of images to the database
for paper in list_of_papers:
    #Initializing variable to indicate start row for adding data to xlsx
    k=1
    # some time a .DS_STORE folder is created. we need to disread that folder
    if paper != ".DS_Store":
        # Entering sub directory for each paper. This sub directory contains all the images from given paper.
        os.chdir(paper)
        if os.path.exists(paper+".xlsx"):
            os.system("rm -rf "+ paper + ".xlsx")
        #Create an new database file.
        workbook = xlsxwriter.Workbook(paper+".xlsx")
        # Create a new worksheet for each folder representing a unique paper.
        worksheet = workbook.add_worksheet()
        # The column width should be set as per the max length of the string in the column.
        # To-Do add a code to determine max width of each column and set the column width accordingly.
        # Refer here https://stackoverflow.com/questions/29463274/simulate-autofit-column-in-xslxwriter
        # Widen the first column to make the text clearer.
        worksheet.set_column("A:A", 70)
        worksheet.write(0,0, "Name of Paper")
        worksheet.write(0,1,"Image")
        # Creating a list of all images in current directory.
        list_of_images = os.listdir("./")
        for image in list_of_images:
            if image != ".DS_Store":
                # Writing the paper name in 1st column
                worksheet.write(k,0, paper)
                # inserting the image in 2nd column
                worksheet.insert_image(k,1, image)
                # Increasing the row number by 50 to insert next data.
                k+=50
        workbook.close()
        os.chdir("../")

os.chdir("../../")
print('Done3')
```
